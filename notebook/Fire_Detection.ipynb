```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(
    data="dataset/data.yaml",
    epochs=80,
    imgsz=640,
    batch=8,
    workers=2,
    patience=20,
    lr0=0.001,
    lrf=0.01,
    mosaic=0.5,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    box=7.5,
    cls=1.0,
    dfl=1.5,
    freeze=10
)
```
