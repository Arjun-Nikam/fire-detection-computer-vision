```python
from ultralytics import YOLO
import cv2
import winsound  # Works on Windows for alarm beep

# Load model
model = YOLO("fire_detection_yolov8.pt")

# Open webcam
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    raise Exception("Webcam not found. Check your camera.")

# Frequency and duration for warning sound
ALARM_FREQ = 2500  # Hz
ALARM_DURATION = 800  # milliseconds

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # FIX MIRROR FLIP: Flip horizontally to make it appear natural
    frame = cv2.flip(frame, 1)

    # Run detection
    results = model(frame)
    annotated_frame = results[0].plot()

    # Check detections for "fire"
    fire_detected = False
    for box in results[0].boxes:
        cls_id = int(box.cls[0])  # class index
        class_name = model.names[cls_id]

        if class_name.lower() == "fire":
            fire_detected = True
            break

    # Play alarm if fire detected
    if fire_detected:
        winsound.Beep(ALARM_FREQ, ALARM_DURATION)

    # Display annotated frame
    cv2.imshow("Fire & Smoke Detection", annotated_frame)

    # Quit key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
